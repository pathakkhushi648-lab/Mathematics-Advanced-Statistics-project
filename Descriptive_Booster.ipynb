```python
import matplotlib.pyplot as plt
import seaborn as sns

skewness = df["Household_Income"].skew()
kurtosis = df["Household_Income"].kurt()
print("Skewness:", skewness)
print("Kurtosis:", kurtosis)

plt.figure(figsize=(12,6))
sns.histplot(
    df["Household_Income"],
    kde=True,
    bins=30
)

plt.title("Distribution of Household Income", fontsize=16)

plt.xlabel("Household Income", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.text(
    0.65,
    0.85,
    f"Skewness = {skewness:.2f}\nKurtosis = {kurtosis:.2f}",
    transform=plt.gca().transAxes,
    fontsize=12,
    bbox=dict(facecolor='white', alpha=0.8)
)
plt.grid(True)
plt.show()
```

The graph visualizes income distribution along with skewness and kurtosis values.

If skewness is positive, the distribution is positively skewed. Kurtosis indicates the peakedness of the distribution.

The Gaussian curve approximates the normal distribution of household income

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

income = df["Household_Income"]

mu = income.mean()
sigma = income.std()
plt.figure(figsize=(12,7))

sns.histplot(
    income,
    bins=25,
    stat='density',
    kde=True,
    edgecolor='black'
)

x = np.linspace(
    income.min(),
    income.max(),
    1000
)

y = norm.pdf(x, mu, sigma)

plt.plot(
    x,
    y,
    linewidth=3,
    label='Gaussian Normal Curve'
)

plt.axvline(
    mu,
    linestyle='--',
    linewidth=2,
    label=f'Mean = {mu:.2f}'
)
plt.title("Gaussian Distribution of Household Income", fontsize=18)

plt.xlabel("Household Income", fontsize=13)

plt.ylabel("Density", fontsize=13)
plt.legend()
plt.grid(True)
plt.show()
```
